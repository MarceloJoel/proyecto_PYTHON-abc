### Documentacion de Funciones (docstrings)

Las cadenas de documentación son cadenas de texto que describen el propósito, los parámetros y el valor de retorno de una función.

```python
def rectangle_area(base, height):
    """
    Calculates the area of a rectangle.
    Args:
        base (float): The base of the rectangle.
        height (float): The height of the rectangle.
    Returns:
        float: The area of the rectangle.
    """
    return base * height
```

### Función con número de Variable de Argumentos

```python
def variable_sum(*numbers):
    total = 0
    for number in numbers:
        total+= number
    return total
```

```python
print(variable_sum(1,2,3))
```

```python
print(variable_sum(4,5,6,7))
```

### Excepciones personalizadas

```python
def function():
    if condition:
        raise Exception("Error description") # Lanza un error

try:
    function()

except Exception as e: # Si ocurre un error, entra aquí en vez de romper el programa.
    print(f"Error:", e)
```

```python
def verificar_edad(edad):

    if edad < 18:
        raise Exception("Eres menor de edad")

    print("Acceso permitido")


try:
    verificar_edad(15)

except Exception as e:
    print("Error:", e)
```

### Input/output

```python
name = input("Ingresa tu nombre:")
age = input("Ingresa tu edad:")

print("Hola, " + name + "!")
print("Tienes " + age + "años.")
```

Si desea trabajar con otros tipos de datos, como enteros o números de coma flotante, debe realizar una conversión explícita utilizando funciones como int() o float().

```python
age = int(input("Ingresa tu edad: "))

if age >= 18:
    print("Eres adulto")
else:
    print("Eres menor de edad")
```

**Leer y escribir archivos**

*Leer archivos:*

```python
file = open("data.txt", "r")
content = file.read()
print(content)
file.close()
```

*Escribir archivos:*

```python
file =open("data.txt", "w")
file.write("Hello, World!")
file.close()
```

Es **importante** cerrar siempre los archivos después de usarlos para liberar recursos del sistema.

*gestionar automáticamente la apertura y el cierre de archivos.*

```python
with open("data.txt", "r") as file:
    content = file.read()
    print(content)
```
