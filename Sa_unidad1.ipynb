```python
counter = 0

while True:
    print(counter)
    counter += 1

    if counter == 5:
        break
```

* Continue

```python
for i in range(10):

    if i % 2 == 0:
        continue    # salta lo que queda del ciclo y pasa a la siguiente iteración
    print(i)
```
